```python
# Import Necessary Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Importing the dataset
df = pd.read_csv('zomato.csv')
df
```

```python
# Removing the 'address' and 'phone' columns

del df['address']
del df['phone']
df
```

```python
# Rename the columns: 'rate' to 'rating', 'approx_cost(for two people)' to 'approx_cost', 'listed_in(type)' to 'type'.

df.rename(columns={'rate':'rating'},inplace=True)
df.rename(columns={'approx_cost(for two people)':'approx_cost'},inplace=True)
df.rename(columns={'listed_in(type)':'type'},inplace=True)
df
```

```python
# Remove rows with null values in the 'name' column

df.dropna(subset=['name'],inplace=True)

# # Handling null values in all other columns

df['online_order'] = df['online_order'].fillna('NA')
df['book_table'] = df['book_table'].fillna('NA')
df['rating'] = df['rating'].fillna(0)
df['votes'] = df['votes'].fillna(0)
df['location'] = df['location'].fillna('NA')
df['rest_type'] = df['rest_type'].fillna('NA')
df['dish_liked'] = df['dish_liked'].fillna('NA')
df['cuisines'] = df['cuisines'].fillna('NA')
df['approx_cost'] = df['approx_cost'].fillna(0)
df['type'] = df['type'].fillna('NA')
df
```

```python
# Droping the duplicates value with keeping the first occurance

df.drop_duplicates(keep='first',inplace=True)
df
```

```python
# Filtering Out Rows Containing 'RATED' or 'Rated' in Specific Columns

df = df[df['name'].str.contains('RATED|Rated') == False]
df = df[df['type'].str.contains('RATED|Rated') == False]
df = df[df['approx_cost'].str.contains('RATED|Rated') == False]
df = df[df['cuisines'].str.contains('RATED|Rated') == False]
df = df[df['dish_liked'].str.contains('RATED|Rated') == False]
df = df[df['rest_type'].str.contains('RATED|Rated') == False]
df = df[df['location'].str.contains('RATED|Rated') == False]
df = df[df['votes'].str.contains('RATED|Rated') == False]
df = df[df['rating'].str.contains('RATED|Rated') == False]
df = df[df['book_table'].str.contains('RATED|Rated') == False]
df
```

```python

# 'online order' column should have only 'Yes' or 'No', remove other values

df = df[df['online_order'].str.contains('Yes|No')==True]

# check for 'rating' column and replace 'NEW', '-' values to 0 and remove '/5'

df.loc[:, 'rating'] = df['rating'].apply(lambda x: float(str(x).split('/')[0]) if '/' in str(x) else 0.0)
df
```

```python
# remove unknown character from dataset

df.loc[:,'name'] = df['name'].str.replace('[Ãx][^A-Za-z]+', '', regex = True)

# Export the dataset

df.to_csv('zomatocleaned.csv', index=False)
df
```

```python
# Database Connection

from sqlalchemy import create_engine, text
db_url = "mysql+pymysql://root:****@localhost/zomato_db"
engine = create_engine(db_url)

# Executing simple test query to check the connection
with engine.connect() as conn:
    result = conn.execute(text("SELECT DATABASE();"))
    print(result.fetchone())
```

```python
from tabulate import tabulate      # Used to get output of SQL query in readable table format.

# Open CSV file then create 'zomato' table in database

df = pd.read_csv("zomatocleaned.csv")
table_name = "zomato"
df.to_sql(name=table_name, con=engine,if_exists="replace", index=False)
print(f"Table '{table_name}' created successfully in MySQL ")

# Query to get First 10 data from table

query = "SELECT * FROM zomato LIMIT 10;"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
#  Top 5 delivery restaurants based on the highest number of votes.

query = "SELECT name, votes, rating FROM zomato WHERE type='Delivery' ORDER BY votes DESC LIMIT 5"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
#  Top 5 delivery restaurants in Banashankari based on the highest ratings.

query = "SELECT name,rating,location,type FROM zomato WHERE location='Banashankari' AND type='Delivery' ORDER BY rating DESC LIMIT 5"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
# Ratings of restaurants in Indiranagar having minimum and maximum approximate cost.

query = "SELECT(SELECT rating FROM zomato WHERE location='Indiranagar' AND approx_cost = (SELECT MIN(approx_cost) " \
"FROM zomato WHERE location='Indiranagar') LIMIT 1) AS rating1, " \
"(SELECT rating FROM zomato WHERE location='Indiranagar' AND approx_cost = (SELECT MAX(approx_cost) " \
"FROM zomato WHERE location='Indiranagar') LIMIT 1) AS rating2"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
#  Total number of votes for restaurants grouped by whether they offer online ordering.

query = "SELECT SUM(votes) AS total_votes,online_order FROM zomato GROUP BY online_order"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
#  top 10 restaurant types based on the number of restaurants, total votes, and average rating.

query = "SELECT type,COUNT(*) AS number_of_resturants, SUM(votes) AS total_votes, AVG(rating) AS avg_rating " \
"FROM zomato WHERE type IS NOT NULL " \
"GROUP BY type ORDER BY total_votes DESC LIMIT 10"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
# Name, most_liked_dish, rating and number of votes of highly rated delivery restaurants allowing online orders.

query = "SELECT name, dish_liked AS most_liked_dish, rating, votes FROM zomato " \
"WHERE type = 'Delivery' AND online_order = 'Yes' " \
"AND rating > (SELECT AVG(rating) FROM zomato WHERE online_order='Yes' AND type='Delivery') " \
"ORDER BY rating DESC, votes DESC LIMIT 1"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
# Highly rated restaurants with ratings above 3 and at least 150 votes that do not offer online orders.

query = "SELECT name, rating, votes, online_order FROM zomato " \
"WHERE online_order = 'No' AND rating > 3 AND votes >= 150 LIMIT 10"
df_sql = pd.read_sql(query, engine)
with engine.connect() as conn:
    result = conn.execute(text(query))
    print(tabulate(result.fetchall(), headers=result.keys(), tablefmt='fancy_grid'))
```

```python
# Open CSV file
df = pd.read_csv('zomatocleaned.csv')

# Visualize the top 10 locations with the highest number of restaurants listed on Zomato.

# It helps identify the most saturated food hubs in the city.

plt.figure(figsize=(12,6))
df['location'].value_counts().head(10).plot(kind='bar', color='skyblue')
plt.title("Top 10 locations with most restaurants")
plt.xlabel('Location')
plt.ylabel('Number of Restaurants')
plt.show()
```

```python
# Visualize the distribution of restaurant ratings.

plt.figure(figsize=(10,5))
sns.histplot(df['rating'], bins=20, kde='True', color='red')
plt.title('Rating Distribution')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
# Most ratings lie between 3.0 and 4.5, showing the general customer satisfaction range.

# The KDE curve shows the density of ratings.
```

```python
# Visualize how many restaurants support online ordering.

# It's useful to understand the adoption of digital food delivery in the market.


plt.figure(figsize=(6,4))
sns.countplot(data=df, x='online_order',palette='Set2')
plt.title('Online Order Availability')
plt.xlabel('Online Order')
plt.ylabel('Count')
plt.show()
```

```python
#Visualize the average rating for each type of restaurant.

# It helps understand which restaurant types are most appreciated by customers.

rest_ratings = df.groupby('rest_type')['rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(12,6))
sns.barplot(x=rest_ratings.values, y=rest_ratings.index, palette='coolwarm')
plt.title('Average Rating by Restaurant Type')
plt.xlabel('Average Rating')
plt.ylabel('Restaurant Type')
plt.show()
```

```python
# Visualize the relationship between the number of votes and the rating.

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='votes', y='rating', alpha=0.6, color='purple')
plt.title('Votes vs Rating')
plt.xlabel('Votes')
plt.ylabel('Rating')
plt.grid=True
plt.show()
```

```python
# Each point represents a restaurant.

# A positive trend may suggest that restaurants with more votes tend to have better ratings.
```

```python
# Visualize the proportion of restaurants that offer online ordering vs those that don't.

# It provides a quick overview of digital service availability.

order_counts = df['online_order'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(order_counts, labels=order_counts.index, autopct='%1.1f%%', colors=['red','green'])
plt.title('Online Order Availability')
plt.axis('equal')
plt.show()
```
